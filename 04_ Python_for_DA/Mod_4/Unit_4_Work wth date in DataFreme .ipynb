# Выделение атрибутов datetime

> Тип данных __datetime__ позволяет с помощью специального ___аксессора dt__ выделять составляющие времени из каждого элемента столбца, такие как:
- __date__ — дата;
- __year, month, day__ — год, месяц, день;
- __time__ — время;
- __hour, minute, second__ — час, минута, секунда;
- __dayofweek__ — номер дня недели, от 0 до 6, где 0 — понедельник, 6 — воскресенье;
- __day_name__ — название дня недели;
- __dayofyear__ — порядковый день года;
- __quarter__ — квартал (интервал в три месяца).

```python
sold_year = melb_df_dt['Date'].dt.year
melb_df_dt['YearSold'] = sold_year
print(sold_year)
print(f'Min sold year {sold_year.min()}')
print(f'Max sold year {sold_year.max()}')
print(f'Mode sold year {sold_year.mode()[0]}')
```

```python
melb_df_dt
```

```python
sold_month_num = melb_df_dt['Date'].dt.month
print(sold_month_num)
print(f'Mode sold month {sold_month_num.mode()[0]}')
```

```python
melb_df_dt['MonthSold'] = sold_month_num
melb_df_dt
```

```python
melb_df_dt['MonthSold'].value_counts(normalize=True)
```

```python
sold_month_str = melb_df_dt['Date'].dt.month_name()
print(sold_month_str)
print(f'Mode sold month {sold_month_str.mode()[0]}')
```

```python
sold_day = melb_df_dt['Date'].dt.day
print(sold_day)
print(f'Mode sold day {sold_day.mode()[0]}')
```

# Самостоятельные работы из Unit_5

```python
def get_weekend(weekday):
    weekday_num = [5, 6]
    if weekday in weekday_num:
        return 1
    else:
        return 0
    
melb_df_dt['Weekday'] = melb_df_dt['WeekdaySale'].apply(get_weekend)
melb_df_dt[['WeekdaySale', 'Weekday']]
```

```python
melb_df_dt[melb_df_dt['Weekday'] == 1]['Price'].mean()
```

```python
melb_df_dt['SellerG'].nunique()
```

```python
melb_df_dt['SellerG'].value_counts()
```

```python
top_49_seller = melb_df_dt['SellerG'].value_counts().nlargest(n=49).index
top_49_seller
```

```python
melb_df_dt['SellerG'] = melb_df_dt['SellerG'].apply(lambda x: x if x in top_49_seller else 'other')
```

```python
nilson_prices = melb_df_dt[(melb_df_dt['SellerG'] == 'Nelson')]['Price'].min()
other_prices = melb_df_dt[(melb_df_dt['SellerG'] == 'other')]['Price'].min()
print(nilson_prices / other_prices)
```
