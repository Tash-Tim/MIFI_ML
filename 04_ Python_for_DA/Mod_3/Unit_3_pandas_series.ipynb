- Series — это упорядоченная изменяемая коллекция объектов, имеющая так называемые ассоциативные метки (индексы). 

```python
import pandas as pd
print(pd.__name__)
```

# Создание Series

```python
countries = pd.Series(
    data=['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'], 
    index=['UK', 'CA', 'US', 'RU', 'UA', 'BY', 'KZ'], 
    name='Counries'
)
print(countries)
```

```python

countries_1 = pd.Series(
    {'UK': 'Англия',
    'CA': 'Канада',
    'US' : 'США',
    'RU': 'Россия',
    'UA': 'Украина',
    'BY': 'Беларусь',
    'KZ': 'Казахстан'},
    name='countries_1'
)
display(countries_1)
```

```python
countries_2 = pd.Series(
    data=['Англия', 'Канада', 'США', 'Россия', 'Украина', 'Беларусь', 'Казахстан'])
display(countries_2)
```

# Доступ к элементам в Series (loc & iLoc)

```python
display(countries.loc['KZ'])
```

```python
display(countries_1.loc[['US', 'RU', 'UA']])
```

```python
display(countries.iloc[1])
```

```python
display(countries_1.iloc[2:6])
```

```python
display(countries_2.loc[0])
```

```python
display(countries_2.loc[3:6])
```

```python
display(countries_2.iloc[0])
```

```python
display(countries_2.iloc[3:6])
```

> На самом деле loc и iloc можно опустить и обращаться к элементам Series напрямую по индексам, например countries[[‘UK’, 'US', ‘UA’]] или countries[[0, 2, 4]]. Оба варианта являются равноправными для Series, однако в дальнейшем мы будем использовать эти операции при обращении к более сложной структуре — DataFrame, а в контексте этой структуры эти варианты уже неравноправны.

```python
def create_medications(names, counts):
    return pd.Series(data=counts, index=names)

def get_percent(medications, name):
    return round((medications[name] * 100) / sum(medications.values), 3)
```

```python
print(create_medications(names = ['chlorhexidine', 'cyntomycin', 'afobazol'], counts = [26, 18, 7]))
```

```python
medications = create_medications(names = ['chlorhexidine', 'cyntomycin', 'afobazol'], counts = [26, 18, 7])
print(get_percent(medications, "chlorhexidine"))
```
