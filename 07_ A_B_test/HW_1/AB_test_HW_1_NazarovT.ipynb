# –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–π –∞–Ω–∞–ª–∏–∑ –¥–≤—É—Ö –±–∏–Ω–∞—Ä–Ω—ã—Ö –≤—ã–±–æ—Ä–æ–∫

## üìä –£—Å–ª–æ–≤–∏–µ –∑–∞–¥–∞—á–∏

### –ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:
- **–î–≤–µ –±–∏–Ω–∞—Ä–Ω—ã–µ –≤—ã–±–æ—Ä–∫–∏** —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—è–º–∏ —É—Å–ø–µ—Ö–∞:
  - $p_1 = 0.35$ (–ø–µ—Ä–≤–∞—è –≤—ã–±–æ—Ä–∫–∞)
  - $p_2 = 0.40$ (–≤—Ç–æ—Ä–∞—è –≤—ã–±–æ—Ä–∫–∞)
- **–£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏**: $\alpha = 0.05$ (p-value)
- **–ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π –¥–µ—Ç–µ–∫—Ç–∏—Ä—É–µ–º—ã–π —ç—Ñ—Ñ–µ–∫—Ç (MDE)**: $\delta = |p_2 - p_1| = 0.05$

### –ü–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–µ –∑–∞–¥–∞—á–∏:

#### 1. üìà –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –≤—ã–±–æ—Ä–∫–∏
**–¶–µ–ª—å**: –ù–∞–π—Ç–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ $n$ –≤ –∫–∞–∂–¥–æ–π –≤—ã–±–æ—Ä–∫–µ, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–µ –¥–ª—è –¥–µ—Ç–µ–∫—Ü–∏–∏ MDE —Å –∑–∞–¥–∞–Ω–Ω–æ–π –º–æ—â–Ω–æ—Å—Ç—å—é —Ç–µ—Å—Ç–∞ ($1 - \beta = 0.8$).

**–§–æ—Ä–º—É–ª–∞ –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞**:
$$
n = \frac{(z_{1-\alpha/2} \cdot \sqrt{2 \cdot \bar{p} \cdot (1-\bar{p})} + z_{1-\beta} \cdot \sqrt{p_1(1-p_1) + p_2(1-p_2)})^2}{(p_1 - p_2)^2}
$$

–≥–¥–µ:
- $\bar{p} = \frac{p_1 + p_2}{2}$ - –æ–±—ä–µ–¥–∏–Ω–µ–Ω–Ω–∞—è –ø—Ä–æ–ø–æ—Ä—Ü–∏—è
- $z_{1-\alpha/2}$ - –∫–≤–∞–Ω—Ç–∏–ª—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–ª—è –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–µ–≥–æ —Ç–µ—Å—Ç–∞
- $z_{1-\beta}$ - –∫–≤–∞–Ω—Ç–∏–ª—å –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –¥–ª—è –º–æ—â–Ω–æ—Å—Ç–∏ —Ç–µ—Å—Ç–∞

#### 2. üîÑ –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π —á–µ—Ä–µ–∑ –¶–ü–¢
**–¶–µ–ª—å**: –ò—Å–ø–æ–ª—å–∑—É—è –¶–µ–Ω—Ç—Ä–∞–ª—å–Ω—É—é –ü—Ä–µ–¥–µ–ª—å–Ω—É—é –¢–µ–æ—Ä–µ–º—É (–¶–ü–¢), –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞—Ç—å –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∫ –Ω–æ—Ä–º–∞–ª—å–Ω—ã–º.

**–§–æ—Ä–º—É–ª—ã –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è**:
- –î–ª—è –≤—ã–±–æ—Ä–∫–∏ —Ä–∞–∑–º–µ—Ä–∞ $n$ —Å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å—é $p$:
  $$
  \text{–°—Ä–µ–¥–Ω–µ–µ: } \mu = p
  $$
  $$
  \text{–°—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ: } \sigma = \sqrt{\frac{p(1-p)}{n}}
  $$
- –ü–æ–ª—É—á–∞–µ–º –Ω–æ—Ä–º–∞–ª—å–Ω—ã–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:
  - –ü–µ—Ä–≤–∞—è –≤—ã–±–æ—Ä–∫–∞: $N(\mu_1, \sigma_1)$
  - –í—Ç–æ—Ä–∞—è –≤—ã–±–æ—Ä–∫–∞: $N(\mu_2, \sigma_2)$

#### 3. üìä –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –∏ –æ—à–∏–±–æ–∫
**–¢—Ä–µ–±—É–µ—Ç—Å—è –ø–æ—Å—Ç—Ä–æ–∏—Ç—å –¥–≤–∞ –≥—Ä–∞—Ñ–∏–∫–∞**:

**–ì—Ä–∞—Ñ–∏–∫ 1**: –î–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω—è—è –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–∞
- –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ–±–µ–∏—Ö –≤—ã–±–æ—Ä–æ–∫
- –ö—Ä–∏—Ç–∏—á–µ—Å–∫–∏–µ –æ–±–ª–∞—Å—Ç–∏ –¥–ª—è –æ—à–∏–±–∫–∏ I —Ä–æ–¥–∞ ($\alpha$)
- –û–±–ª–∞—Å—Ç—å –æ—à–∏–±–∫–∏ II —Ä–æ–¥–∞ ($\beta$)

**–ì—Ä–∞—Ñ–∏–∫ 2**: –û–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω—è—è –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–∞ ($p_1 > p_2$)
- –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –æ–±–µ–∏—Ö –≤—ã–±–æ—Ä–æ–∫  
- –û–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–µ –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–µ –æ–±–ª–∞—Å—Ç–∏
- –°–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏–µ –æ–±–ª–∞—Å—Ç–∏ –æ—à–∏–±–æ–∫

**–û–±–æ–∑–Ω–∞—á–µ–Ω–∏—è –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–∞—Ö**:
- üü¶ **–°–∏–Ω—è—è –æ–±–ª–∞—Å—Ç—å** - –û—à–∏–±–∫–∞ I —Ä–æ–¥–∞ (–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å $\alpha$)
- üü• **–ö—Ä–∞—Å–Ω–∞—è –æ–±–ª–∞—Å—Ç—å** - –û—à–∏–±–∫–∞ II —Ä–æ–¥–∞ (–≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å $\beta$)
- üìç **–í–µ—Ä—Ç–∏–∫–∞–ª—å–Ω—ã–µ –ª–∏–Ω–∏–∏** - –ö—Ä–∏—Ç–∏—á–µ—Å–∫–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è –∏ —Å—Ä–µ–¥–Ω–∏–µ

#### 4. üìê –†–∞—Å—á–µ—Ç –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã—Ö –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤
**–¶–µ–ª—å**: –ü–æ—Å—Ç—Ä–æ–∏—Ç—å $95\%$ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –¥–ª—è —Å—Ä–µ–¥–Ω–∏—Ö –∑–Ω–∞—á–µ–Ω–∏–π –æ–±–µ–∏—Ö –≤—ã–±–æ—Ä–æ–∫.

**–§–æ—Ä–º—É–ª–∞ –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω–æ–≥–æ –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞**:
$$
CI = \hat{p} \pm z_{1-\alpha/2} \cdot \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}
$$

–≥–¥–µ:
- $\hat{p}$ - –≤—ã–±–æ—Ä–æ—á–Ω–∞—è –ø—Ä–æ–ø–æ—Ä—Ü–∏—è
- $z_{1-\alpha/2}$ - –∫–≤–∞–Ω—Ç–∏–ª—å —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è

### üéØ –û–∂–∏–¥–∞–µ–º—ã–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã:

1. **–ß–∏—Å–ª–æ–≤–æ–π —Ä–µ–∑—É–ª—å—Ç–∞—Ç**: –ú–∏–Ω–∏–º–∞–ª—å–Ω—ã–π —Ä–∞–∑–º–µ—Ä –≤—ã–±–æ—Ä–∫–∏ $n$
2. **–ü–∞—Ä–∞–º–µ—Ç—Ä—ã —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π**: $\mu_1, \sigma_1, \mu_2, \sigma_2$
3. **–í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è**: –î–≤–∞ –≥—Ä–∞—Ñ–∏–∫–∞ —Å –æ–±–ª–∞—Å—Ç—è–º–∏ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏—Ö –æ—à–∏–±–æ–∫
4. **–ò–Ω—Ç–µ—Ä–≤–∞–ª—å–Ω—ã–µ –æ—Ü–µ–Ω–∫–∏**: –î–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã–µ –∏–Ω—Ç–µ—Ä–≤–∞–ª—ã –¥–ª—è $p_1$ –∏ $p_2$

### üìà –°—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –≥–∏–ø–æ—Ç–µ–∑—ã:

**–î–ª—è –¥–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–µ–≥–æ —Ç–µ—Å—Ç–∞**:
$$
H_0: p_1 = p_2 \\
H_1: p_1 \neq p_2
$$

**–î–ª—è –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–µ–≥–æ —Ç–µ—Å—Ç–∞**:
$$
H_0: p_1 \leq p_2 \\
H_1: p_1 > p_2
$$

```python
# –í–∞—à –∫–æ–¥
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

print('=' * 70)
print('–ê–ù–ê–õ–ò–ó –î–í–£–• –ë–ò–ù–ê–†–ù–´–• –í–´–ë–û–†–û–ö (–° –ù–ï–†–ê–í–ù–´–ú–ò –†–ê–ó–ú–ï–†–ê–ú–ò)')
print('=' * 70)

print('''–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤ –≥—Ä—É–ø–ø–µ 1: 0.35
–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –≤ –≥—Ä—É–ø–ø–µ 2: 0.4
–£—Ä–æ–≤–µ–Ω—å –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ (p-value): 0.05
–ú–æ—â–Ω–æ—Å—Ç—å —Ç–µ—Å—Ç–∞: 0.8''')
print('-' * 70)


#–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
p1 = 0.35
p2 = 0.40
alpha = 0.05
power = 0.8


# –ó–∞–¥–∞—á–∞ 1. –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞ –≤—ã–±–æ—Ä–∫–∏
p_bar = (p1 + p2) / 2
z_alpha = norm.ppf(1 - alpha / 2)
z_power = norm.ppf(power)
n = ((z_alpha * np.sqrt(2 * p_bar * (1 - p_bar)) +z_power * np.sqrt(p1 * (1 - p1) + p2 * (1 - p2))) ** 2) / (p1 - p2) ** 2
n = int(np.round(n))

print('\n1. –ú–ò–ù–ò–ú–ê–õ–¨–ù–´–ô –†–ê–ó–ú–ï–† –í–´–ë–û–†–ö–ò (–†–ê–í–ù–´–ï –ì–†–£–ü–ü–´):')
print(f'    n‚ÇÅ = n‚ÇÇ = {n} —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤ –∫–∞–∂–¥–æ–π –≥—Ä—É–ø–ø–µ')


# –ó–∞–¥–∞—á–∞ 2. –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π —á–µ—Ä–µ–∑ –¶–ü–¢
    #—Ä–∞–≤–Ω—ã–µ –≤—ã–±–æ—Ä–∫–∏ n=1471
sigma_1 = p1 * (1 - p1) / n
sigma_2 = p2 * (1 - p2) / n
sigma1_eq = np.sqrt(sigma_1)
sigma2_eq = np.sqrt(sigma_2)
    #–Ω–µ—Ä–∞–≤–Ω—ã–µ –≤—ã–±–æ—Ä–∫–∏
n1 = 735
n2 = 7350
sigma1_uneq = np.sqrt(p1 * (1 - p1) / n1)
sigma2_uneq = np.sqrt(p2 * (1 - p2) / n2)

print('\n2. –ü–†–ï–û–ë–†–ê–ó–û–í–ê–ù–ò–ï –ß–ï–†–ï–ó –¶–ü–¢:')
print(f'    –†–∞–≤–Ω—ã–µ –≤—ã–±–æ—Ä–∫–∏ (n={n}):')
print(f'        –ì—Ä—É–ø–ø–∞ 1: N(Œº={p1:.4f}, œÉ={sigma1_eq:.4f})')
print(f'        –ì—Ä—É–ø–ø–∞ 1: N(Œº={p2:.4f}, œÉ={sigma2_eq:.4f})')
print(f'    –ù–µ—Ä–∞–≤–Ω—ã–µ –≤—ã–±–æ—Ä–∫–∏ (n‚ÇÅ={n1}, n‚ÇÇ={n2}):')
print(f'        –ì—Ä—É–ø–ø–∞ 1: N(Œº={p1:.4f}, œÉ={sigma1_uneq:.4f})')
print(f'        –ì—Ä—É–ø–ø–∞ 1: N(Œº={p2:.4f}, œÉ={sigma2_uneq:.4f})')


# –ó–∞–¥–∞—á–∞ 3. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π –∏ –æ—à–∏–±–æ–∫
print('\n3. –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø –†–ê–°–ü–†–ï–î–ï–õ–ï–ù–ò–Ø –ò –û–®–ò–ë–ö–û–ö')
    # —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ä–∞–∑–Ω–æ—Å—Ç–∏ –¥–æ–ª–µ–π
mu0 = 0          # H0: p1 = p2
mu1 = p2 - p1    # H1: p2 - p1
sigma_diff = np.sqrt(sigma_1 + sigma_2)
x = np.linspace(-0.05, 0.12, 2000)
pdf_H0 = norm.pdf(x, mu0, sigma_diff)
pdf_H1 = norm.pdf(x, mu1, sigma_diff)
    # –¥–≤—É—Ö—Å—Ç–æ—Ä–æ–Ω–Ω—è—è –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–∞
crit_left = -z_alpha * sigma_diff
crit_right = z_alpha * sigma_diff
print('     –î–≤—É—Ö—Å—Ç–æ—Ä–æ–Ω–Ω—è—è –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–∞')
plt.figure(figsize=(10, 5))
plt.plot(x, pdf_H0, label='H‚ÇÄ: p‚ÇÅ = p‚ÇÇ')
plt.plot(x, pdf_H1, label='H‚ÇÅ: p‚ÇÅ ‚â† p‚ÇÇ')
        # –û—à–∏–±–∫–∞ I —Ä–æ–¥–∞ (Œ±)
plt.fill_between(x, pdf_H0, where=(x <= crit_left), alpha=0.3)
plt.fill_between(x, pdf_H0, where=(x >= crit_right), alpha=0.3)
        # –û—à–∏–±–∫–∞ II —Ä–æ–¥–∞ (Œ≤)
plt.fill_between(x, pdf_H1, where=(x >= crit_left) & (x <= crit_right), color='red', alpha=0.3)
plt.axvline(crit_left, linestyle='--')
plt.axvline(crit_right, linestyle='--')
plt.axvline(mu0, linestyle=':')
plt.axvline(mu1, linestyle=':')
plt.title('–î–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π —Ç–µ—Å—Ç: –æ—à–∏–±–∫–∏ I –∏ II —Ä–æ–¥–∞')
plt.legend()
plt.show()
    # –æ–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω—è—è –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–∞
z_one = norm.ppf(1 - alpha)
crit_one = z_one * sigma_diff
print('     –û–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω—è—è –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–∞')
plt.figure(figsize=(10, 5))
plt.plot(x, pdf_H0, label='H‚ÇÄ: p‚ÇÅ ‚â§ p‚ÇÇ')
plt.plot(x, pdf_H1, label='H‚ÇÅ: p‚ÇÅ > p‚ÇÇ')
        # –û—à–∏–±–∫–∞ I —Ä–æ–¥–∞
plt.fill_between(x, pdf_H0, where=(x >= crit_one), alpha=0.3)
        # –û—à–∏–±–∫–∞ II —Ä–æ–¥–∞
plt.fill_between(x, pdf_H1, where=(x <= crit_one),color='red', alpha=0.3)
plt.axvline(crit_one, linestyle='--')
plt.axvline(mu0, linestyle=':')
plt.axvline(mu1, linestyle=':')
plt.title('–û–¥–Ω–æ—Å—Ç–æ—Ä–æ–Ω–Ω–∏–π —Ç–µ—Å—Ç: –æ—à–∏–±–∫–∏ I –∏ II —Ä–æ–¥–∞')
plt.legend()
plt.show()


#–ó–∞–¥–∞—á–∞ 4. –†–∞—Å—á–µ—Ç –¥–æ–≤–µ—Ä–∏—Ç–µ–ª—å–Ω—ã—Ö –∏–Ω—Ç–µ—Ä–≤–∞–ª–æ–≤
ci1_eq = (p1 - z_alpha * sigma1_eq, p1 + z_alpha * sigma1_eq )
ci2_eq = (p2 - z_alpha * sigma2_eq, p2 + z_alpha * sigma2_eq)

print('\n4. –î–û–í–ï–†–ò–¢–ï–õ–¨–ù–´–ï –ò–ù–¢–ï–†–í–ê–õ–´ (—É—Ä–æ–≤–µ–Ω—å –¥–æ–≤–µ—Ä–∏—è 0.95):')
print('\n   –†–ê–í–ù–´–ï –í–´–ë–û–†–ö–ò:')
print(f'     –ì—Ä—É–ø–ø–∞ 1: [{ci1_eq[0]:.4f}, {ci1_eq[1]:.4f}] —à–∏—Ä–∏–Ω–∞: {ci1_eq[1] - ci1_eq[0]:.4f})')
print(f'     –ì—Ä—É–ø–ø–∞ 2: [{ci2_eq[0]:.4f}, {ci2_eq[1]:.4f}] —à–∏—Ä–∏–Ω–∞: {ci2_eq[1] - ci2_eq[0]:.4f})')

ci1_uneq = (p1 - z_alpha * sigma1_uneq, p1 + z_alpha * sigma1_uneq)
ci2_uneq = ( p2 - z_alpha * sigma2_uneq, p2 + z_alpha * sigma2_uneq)

print('\n   –ù–ï–†–ê–í–ù–´–ï –í–´–ë–û–†–ö–ò:')
print(f'     –ì—Ä—É–ø–ø–∞ 1: [{ci1_uneq[0]:.4f}, {ci1_uneq[1]:.4f}] —à–∏—Ä–∏–Ω–∞: {ci1_uneq[1] - ci1_uneq[0]:.4f})')
print(f'     –ì—Ä—É–ø–ø–∞ 2: [{ci2_uneq[0]:.4f}, {ci2_uneq[1]:.4f}] —à–∏—Ä–∏–Ω–∞: {ci2_uneq[1] - ci2_uneq[0]:.4f})')
```
