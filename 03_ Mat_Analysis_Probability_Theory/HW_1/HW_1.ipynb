**Задача 1** (14 баллов)

Найдите экстремумы функции:

$$
f(x, y, z)=2 x^3+2 x y+2 x z+y^2+z^2+2 y-8 .
$$

Распишите подробное решение.

```python
from sympy import symbols, diff, solve, Eq, Matrix
import numpy as np
# объявляем все необходимые переменные и функцию
x, y, z = symbols('x, y, z', real=True)
f = 2*x**3 + 2*x*y + 2*x*z + y**2 + z**2 + 2*y - 8

# вычиляем все частные поизводные
x_diff = diff(f, x)
y_diff = diff(f, y)
z_diff = diff(f, z)
print(f'Система уравнений (первые частные производные):\n{x_diff} = 0\n{y_diff} = 0\n{z_diff} = 0\n')

# решаем систему уравений, вычисляем стационарные точки
stat_point = solve((x_diff, y_diff, z_diff), x, y ,z, dict=True)
print(f'Стационарные точки {stat_point}\n')

# Вычисляем все частные произврдные 2-ого порядка
# 1-й метод
hes_matrix = Matrix(([diff(f, x, x), diff(f, x, y), diff(f, x, z)], 
                    [diff(f, y, x), diff(f, y, y), diff(f, y, z)],
                    [diff(f, z, x), diff(f, z, y), diff(f, z, z)]))
print(f'Матрица Гесса (вторые частные прогизводные):\n{np.array(hes_matrix)}\n')


# подставляем стационарные точки в матрицу Гесса и вычислим собственные числа
print(f'Подставим стационарные точки в матрицу Гесса и вычислим собственные значения:')
for i_point in stat_point:
    hes_matr_num = hes_matrix.subs({x:i_point[x], y:i_point[y], z:i_point[z]})
    eigs = [float(i_eig) for i_eig in hes_matr_num.eigenvals().keys()]
    
    print(f'\nСтационарные точки:\n{i_point}\nМатрица Гессе:\n{np.array(hes_matr_num)}')
    print(f'Cобственные значения матрицы Гессе:\n{eigs}')
    
    if all(i>0 for i in eigs):
        print(f'ВЫВОД: точки {i_point} - точка МИНИМУМА')
    elif all(i<0 for i in eigs):
        print(f'ВЫВОД: точки {i_point} - точка МАКСИМУМА')
    else:
        print(f'ВЫВОД: точки {i_point} - СЕДЛОВАЯ точка')
```

```python
# объявляем все необходимые переменные, целевую функцию и условие 
x, y, l = symbols('x, y, l', real = True)
f_maim = 4*x + 8*y
f_cond = y**2 - 2*x*y + 5
f_lagran = f_maim + l * f_cond

# вычиляем все частные поизводные
x_diff = diff(f_lagran, x)
y_diff = diff(f_lagran, y)
l_diff = diff(f_lagran, l)
print(f'Система уравнений (первые частные производные):\n{x_diff} = 0\n{y_diff} = 0\n{l_diff} = 0\n')

# решаем систему уравений, вычисляем стационарные точки
stat_pnt = solve([x_diff, y_diff, l_diff], x, y, l, dict=True)
print(f'Стационарные точки {stat_pnt}\n')

# подставим собственные числа в целевую функцию и расчитаем значение функции
for i_pnt in stat_pnt:
    f_main_result = f_maim.subs({x: i_pnt[x], y: i_pnt[y]})
    print(f'f(x,y) = {f_main_result}')
    if f_main_result < 0:
        print(f'в точке x:{i_pnt[x]}, y:{i_pnt[y]} - целефая функция достигает точки МИНИМУМА\n')
    else:
        print(f'точке x:{i_pnt[x]}, y:{i_pnt[y]} - целефая функция достигает точки МАКСИМУМА\n')
```

```python
def cost_function(X, y , par):
    y_pred = np.dot(X, par)
    error = (y_pred - y)**2
    cost = 1/(n)*np.sum(error)
    return cost

def grad_d(X, y, par, alpha, iterations):
    costs = []
    for i in range(iterations):
        y_pred = np.dot(X, par)
        der = np.dot (X.transpose(), (y_pred - y))/ n
        par -= alpha * der
        costs.append(cost_function(X, y, par))
    return par, costs
```
