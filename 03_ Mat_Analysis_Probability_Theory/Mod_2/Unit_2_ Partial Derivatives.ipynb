# Частные производные

```python
from sympy import symbols, diff, sin

a, b, x = symbols('a b x', real=True)
f = 2 * a * x + b ** 2
print(diff(f, a), diff(f, b), diff(f, x))
```

```python
print(diff(f, a, 2), diff(f, b, 2), diff(f, x, 2))
```

```python
f = (5 - a - 2*b)**2
print(diff(f, a))
print(diff(f, b))
```

```python
x, y = symbols('x y', real=True)
f = x**2 + 3*x*y + 4*y**2
print(diff(f, x))
print(diff(f, y))
```

```python
f = y*sin(x) + sin(y)
print(diff(f, x))
print(diff(f, y))
```
