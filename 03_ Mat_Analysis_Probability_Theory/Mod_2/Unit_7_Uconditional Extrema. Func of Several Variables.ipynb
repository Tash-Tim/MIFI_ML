## Найти матрицу Гессе

для функции $$ f(x, y) = e^x + 5y^3x $$

```python
import numpy as np
import pandas as pd
from sympy import *
x, y = symbols('x, y', real=True)
f = exp(x) + 5*y**3*x
matrix = pd.DataFrame(([diff(f, x, 2), diff(f, x, y)], [diff(f, y, x), diff(f, y, 2)]))
print(matrix)
```
