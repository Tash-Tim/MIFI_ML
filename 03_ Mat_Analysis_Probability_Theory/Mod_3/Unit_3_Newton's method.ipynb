# Метод Ньютона  

Метод Ньютона изначально появился как метод решения уравнений вида
$$
f(x)=0

1. Для одномерных случаев
$$
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
$$
2. Для мномерных случаев
$$
x^{(n+1)} = x^{(n)} - \left[ Hf\!\left(x^{(n)}\right) \right]^{-1} \nabla f\!\left(x^{(n)}\right)

# Самостоятельные работы модуля

- Найдите третий корень полинома 
$$
f(x) = 6x^5 - 5x^4 - 4x^3 + 3x^2
$$

```python
from sympy import Symbol, diff, solve
x = Symbol('x')
f = 6*x**5 - 5*x**4 - 4*x**3 + 3*x**2
f_diff = diff(f, x)
x_st = 0.7
for i_x in range(50):
    f_solv = f.subs({x:x_st})
    f_diff_solv = f_diff.subs({x:x_st})
    newton = x_st - f_solv/f_diff_solv

    print(newton)
    
    x_st= newton
    
    
```

Дана функция.
$$
f(x) = x^3 - 72x - 220
$$
Найдите решение уравнения $f(x) = 0$ для поиска корня в окрестностях точки $x_0 = 12$. 

```python
from sympy import Symbol, diff, solve
x = Symbol('x')
f = x**3 - 72*x - 220
f_diff = diff(f, x)
x_st = 12
for i_x in range(10):
    f_solv = f.subs({x:x_st})
    f_diff_solv = f_diff.subs({x:x_st})
    
    
    newton = x_st - f_solv/f_diff_solv

    print(float(newton))
    
    x_st= newton
```

Найдите положительный корень для уравнения.
$$
x^2 + 9x - 5 = 0
$$
В качестве стартовой точки возьмите $x_0=2.2$.

```python
from sympy import Symbol, diff, solve
x = Symbol('x')
f = x**2 + 9*x - 5
f_diff = diff(f, x)
x_st = 2.2
for i_x in range(10):
    f_solv = f.subs({x:x_st})
    f_diff_solv = f_diff.subs({x:x_st})
    
    newton = x_st - f_solv/f_diff_solv

    print(float(newton))
    
    x_st= newton
```

С помощью метода Ньютона найдите точку минимума для функции.
$$
f(x) = 8x^3 - 2x^2 - 450
$$
В качестве стартовой точки возьмите 42, точность примите за 0.0001.

```python
f = 8*x**3 - 2*x**2 - 450
f_d_1 = diff(f, x)
f_d_2 = diff(f, x, 2)
x_0 = 42
epsilon = 0.0001
f_d_1_solv = f_d_1.subs({x : x_0})

while f_d_1_solv > epsilon:
    f_d_1_solv = f_d_1.subs({x : x_0})
    f_d_2_solv = f_d_2.subs({x : x_0})
    x_curr = x_0 - f_d_1_solv / f_d_2_solv
    print(float(x_curr))
    x_0 = x_curr
```
