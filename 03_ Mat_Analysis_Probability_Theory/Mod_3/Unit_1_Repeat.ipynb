Пусть прибыль в вашей компании выражается следующей функцией, которая зависит от параметра x — количества производимых товаров:
$$
f(x) = -x^4 + 6x^3 - 4x^2 + 80
$$
Найдите максимально возможную прибыль, которую вы можете получить, варьируя количество произведённых товаров.

```python
from sympy import *
x = Symbol('x')
f = -x**4 + 6*x**3 - 4*x**2 + 80
f_diff = diff(f, x)
eq_diff = Eq(f_diff, 0)
result = solve(eq_diff, x)

print(f_diff)
print(result, type(result))
```

```python
# максимально возможную прибыль
for i in result:
    prib = f.subs(x, i)
    print(f' max прибыль {float(prib)}')
```

Допустим, вы хотите произвести некоторое количество товара, которое зависит от часов работы двух ключевых сотрудников следующим образом:
$$
f(x, y) = x^2 + 2y^2
$$
где  — время работы 1-ого сотрудника,  — время работы 2-ого сотрудника.  
Однако вы должны оплатить этим сотрудникам ровно 20 часов работы в сумме, то есть:
$$
x + y = 20
$$
Какое наибольшее количество товаров вы сможете произвести в таком случае?

```python
x, y, l = symbols('x, y, l', real=True)
f = x**2 + 2*y**2
f_lim = x + y - 20
f_lagra = f + l * f_lim

x_diff = diff(f_lagra, x)
y_diff = diff(f_lagra, y)
l_diff = diff(f_lagra, l)

solv = solve([x_diff, y_diff, l_diff], x, y, l)
print(solv)
print(round(float(solv[x])))
print(round(float(solv[y])))

resul = f_lagra.subs(x, round(float(solv[x]))).subs(y, round(float(solv[y]))).subs(l, round(float(solv[l])))
print(resul)
```
