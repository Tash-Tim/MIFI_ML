```python
#reverse_words
read_file = ['Привет как дела', 'На улице идет дождь', 'Я люблю программирование']
print(f'Текст до обработки:')
for i_text in read_file:
    print(i_text)
rev_sent = [' '.join(reversed(i_sent.split())) for i_sent in read_file]
blocks = [", ".join((rev_sent[i], rev_sent[(i+1) % len(rev_sent)])) for i in range(len(rev_sent))]
print(f'\nТекст после обработки:\n {'\n '.join(blocks)}')
```

```python
def str_comres(get_str:str) -> str:
    result = ''
    sym_cnt = {i_unq:get_str.count(i_unq) for i_unq in get_str}
    for i_sym, i_cnt in sym_cnt.items():
        if i_cnt > 1:
            result += i_sym + str(i_cnt)
        else:
            result += i_sym

    return result


#user_str = input('Введите строку: ')
user_str = ['affbbbfghaaag', 'asdrrrhhhjdd', 'qwerty', 'assdddffff']
for i_str in user_str:
    print(f'Исходный текст: {i_str}')
    print(f'Сжатая строка: {str_comres(i_str)}\n')
```

```python
def bracket_check(text: str) -> bool:
    open_brack = {'(', '[', '{'}
    close_brack = {')': '(', ']': '[', '}': '{'}
    open_brack_lst = []

    for i_text in text:
        if i_text in open_brack:
            open_brack_lst.append(i_text)
        elif i_text in close_brack:
            if not open_brack_lst or open_brack_lst[-1] != close_brack[i_text]:
                return False
            open_brack_lst.pop()
    return not open_brack_lst

user_text = ['()','{}','{[[]]}','({(]}',')(','([{}])']
for i_bracket in user_text:
    print(f'{i_bracket} - {bracket_check(i_bracket)}')
```

```python
import random
import string

def pas_gen(pas_length :int) -> str:
    all_sym = string.ascii_letters + string.digits + string.punctuation
    # вместо библиотеки string можно вручную составить текст из всех цифр, букв и символов

    frst_sym = random.choice(string.ascii_letters + string.digits)
    res_sym = ''.join([random.choice(all_sym) for _ in range(pas_length - 1)])

    result = frst_sym + res_sym
    return result


user_length = [5, 8, 10, 12, 16]
for i_len in user_length:
    print('Пароль сгенерирован:', pas_gen(i_len))
```

```python
def about_palin_fun(text: str):

    def text_modif_fun(get_text):
         for i_index in range(len(get_text)):
            mod_text = get_text[: i_index] + get_text[i_index + 1:]
            if mod_text == mod_text[::-1]:
                return f'{get_text} - палиндром, если удалить {get_text[i_index]}'
         else:
            return f'{get_text} - НЕ может быть палиндромом'
             
    check_text = text[::-1]
    if text == check_text:
        return f'{text} - палиндром'
    else:
        return text_modif_fun(text)


user_text = ['мадам', 'qwertewq', 'папа', 'samson', 'samsung']
for i_text in user_text:   
    print(about_palin_fun(i_text))
```
